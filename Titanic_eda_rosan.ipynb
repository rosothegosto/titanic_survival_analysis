```python
import pandas as pd
df = pd.read_csv('train.csv')
df.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.heatmap(df.isnull(), cbar=False, cmap='viridis')
plt.title("Missing data visualization")
plt.show()
```

```python
df['Age']=df['Age'].fillna(df['Age'].median())
```

```python
df['Embarked']=df['Embarked'].fillna(df['Embarked'].mode()[0])
```

```python
df = df.drop('Cabin', axis=1)
```

```python
sns.countplot(x='Sex', hue='Survived', data=df)
```

```python
df.head()
```

```python
ax = sns.countplot(x='Sex', hue='Survived', data=df)
ax.set_title('Sex vs Survival')
ax.set_xlabel('Sex')
ax.set_ylabel('No of People')
ax.legend(title='Survived?')
```

```python
sns.countplot(x='Pclass', hue='Survived', data=df)
```

```python
ax2 = sns.countplot(x='Pclass', hue='Survived', data=df)
ax2.set_title('Class vs survival')
ax2.set_xlabel('Class')
ax2.set_ylabel('No of People')
ax2.legend(title='Survived?')
```

```python
ax3=sns.histplot(data=df, x='Age', bins=20, kde=True, color='skyblue')
ax3.set_title('Age distribution')
ax3.set_xlabel('Age')
ax3.set_ylabel('No of People')
```

```python
sns.boxplot(x='Survived', y='Age', data=df)
```
